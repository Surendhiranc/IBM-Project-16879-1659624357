# 6. Find the outliers and replace the outliers

```python
sns.boxplot(data['Diameter'])
```

```python
data.mean()
```

```python
data.std()
```

```python
q=data.quantile(q=[0.75,0.5])
iqr=q.iloc[0]-q.iloc[1]
iqr
```

```python
l=q.iloc[1]-(1.5*iqr)
l['Diameter']
```

```python
u=q.iloc[1]+(1.5*iqr)
u['Diameter']
```

```python
sns.boxplot(data['Diameter'])
```

```python
data['Diameter']=np.where(data['Diameter']>u['Diameter'],u['Diameter'],np.where(data['Diameter']<l['Diameter'],l['Diameter'],data['Diameter']))
sns.boxplot(data['Diameter'])
```

# 7.Check for Categorical columns and perform encoding

```python
data['Sex'].replace({'F':1,'M':0},inplace=True)
```

```python
data.head()
```

# 8.Spilt Data int Dependent and Independent Variables

```python
y=data['Sex']
y
```

```python
x=data.drop(columns=['Sex'],axis=1)
x.head()
```

# 9. Scalling

```python
names=x.columns
names
```

```python
from sklearn.preprocessing import scale
```

```python
x = scale(x)
```

```python
x
```

# 10.Split the data into Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
print(x_train.shape,x_test.shape)
```

# 11. Build the Model


```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10,criterion='entropy')
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict_train = model.predict(x_train)
```

# 12. Train the Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print('Training accuracy: ',accuracy_score(y_train,y_predict_train))
```

# 13. Test the Model

```python
print('Testing accuracy: ',accuracy_score(y_test,y_predict))
```

# 14. Measure the performance using Metrics


```python
pd.crosstab(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
